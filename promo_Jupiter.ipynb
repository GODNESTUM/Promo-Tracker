```python
columns = ["Title","Partner_Name","Added_by"]
values = ["M","Steam","Anas"]
def filt(columns,values):
    qry = ""
    s = 0
    for x, v in zip(columns,values):
        if s != len(columns)-1:
            qry = qry + x + ' == ' + '"'+str(v)+'"' +' & '
        else:
            qry = qry + x + ' == ' +'"'+ str(v)+'"'
        s+=1
    return (qry)
        
x=df.query(filt(columns,values))
print(x)
```
